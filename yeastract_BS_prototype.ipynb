```python
def FASTA_requester(u): 
    url = requests.post(u)
    return url.text
```

```python
def FASTA_parser(x): 
    
    fasta = FASTA_requester(x)
    
    fasta_split = fasta.split("\n")
    
    if fasta_split[1].startswith(">"): 
        del fasta_split[0]
    
    names = fasta_split[0].replace(">", "").split()
    
    try: 
        sysname = names[0]
    except: 
        sysname = " "
    
    try: 
        protname = names[1]
    except: 
        protname = " "
    
    try: 
        protein = "".join(fasta_split[1:])
    except: 
        protein = " "
    
    return (sysname, protname, protein)
```
